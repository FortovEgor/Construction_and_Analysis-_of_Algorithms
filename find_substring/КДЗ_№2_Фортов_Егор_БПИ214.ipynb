# Таблица №1 (подстроки НЕ содержат символа подстановки '?')

```python
import pandas as pd
 
# Читаем данные из .csv файла и сохраняем их в датафрейм
df = pd.read_csv("table_1_new.csv", ";", index_col=False)

# df.head()
display(df)
# print(df)
display(dict(df.iloc[0])) # // вывод первой строки датафрейма
```

```python
# График №1:
# Вид строки: Случайно сгенерированнный текст в бинарном алфавите - 10'000 символов
# 4 кривых на данном графике (каждая отвечает за определенный алгоритм)
import matplotlib.pyplot as plt 

algo_names = ["Наивный Алгоритм", "КМП стандарт", "КМП утонченный", "Z-функция"]

colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black', 
          'tomato', 'orange', 'darkgreen', 'peru', 'pink', 'darkviolet']

# print(dict(df.iloc[1]))
# for j in range(4):
#   print(dict(df.iloc[0])[algo_names[j]])

x = []  # размеры подстрок (CONST)
for i in range(100, 3001, 100):
  x += [i]

for j in range(4):
  y = []  # время исполнения алгоритма в элементарных операциях
  for i in range(30):
    y.append(dict(df.iloc[i])[algo_names[j]])
  print("J:", j, "y:", y)

  plt.plot(x, y, color=colors[j], label=algo_names[j]) 
  plt.title(dict(df.iloc[0])['Тип шаблона'], fontweight='bold')
  # plt.axis([50, 3100, 0, 70000]) 
  plt.xlabel('Размер подстроки', fontweight='bold')
  plt.ylabel('Время работы алгоритма (Операции)', fontweight='bold')
  plt.legend(loc='best', bbox_to_anchor=(1,1))

plt.show()


# КОММЕНТАРИЙ (интересное наблюдение):
# Наблюдение 1:
# Z-функция имеет большую константу, из-за чего работает она дольше других алгоритмов
# Наблюдение 2:
# Наивный алгоритм ведет себя странно: при бОльших данных он рабоатет чуть быстрее
# Объясняется это тем, что, во-первых, при росте размера паттерна поиска не растет размер основной строки
# во-вторых, алфавит бинарный => очень часто при сравнении паатерна поиска будет раньше происходить break, 
# что приводит к меньшему числу дальнейших сравнений
# и в-третьих, размеры подстроки и строки сильно отличаются (100-3000 VS 10К/100К) 
# Наблюдение 3:
# Сильные скачки (например, в обоих реализациях алгоритма КМП) объяснются тем, что 
# У нас критически малая мощность входного алфавита (0 и 1) (в отличие от Графика №3)
```

```python
# График №2:
# Вид строки: Случайно сгенерированнный текст в бинарном алфавите - 100'000 символов
# 4 кривых на данном графике (каждая отвечает за определенный алгоритм)
import matplotlib.pyplot as plt 

for j in range(4):
  y = []  # время исполнения алгоритма в операциях
  for i in range(30):
    y.append(dict(df.iloc[i + 30])[algo_names[j]])
  print("J:", j, "y:", y)

  plt.plot(x, y, color=colors[j], label=algo_names[j]) 
  plt.title(dict(df.iloc[31])['Тип шаблона'], fontweight='bold')
  # plt.axis([50, 3100, 0, 30000000]) 
  plt.xlabel('Размер подстроки', fontweight='bold')
  plt.ylabel('Время работы алгоритма (Операции)', fontweight='bold')
  plt.legend(loc='best', bbox_to_anchor=(1,1))

plt.show()


# КОММЕНТАРИЙ (интересное наблюдение):
# Наблюдение:
# График наивного алгоритма вырождается в горизонтальную линию -
# - это обяъсняется тем, что размер данных существенно вырос
```
